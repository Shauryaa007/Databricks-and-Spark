By default it will put null in place of corrupted data and create a separate col _corrupted_records to store corrupted records.

%md
In JSON all the non corrupted and corrupted values of any record that is present in _corrupt_record will bw null.

In CSV only the corrupted values of a records will be null.

```python
df = spark.read.format('json').load("/Volumes/test/test_schema/test_volume/sales_corrupted.json")
df.display()
```

```python
df = spark.read.format('json').option('mode','failfast').load("/Volumes/test/test_schema/test_volume/sales_corrupted.json")
df.display()
```

dropmalformed - this mode drops the corrupted records.

```python
df = spark.read.format('json').option('mode','dropmalformed').load("/Volumes/test/test_schema/test_volume/sales_corrupted.json")
df.display()
```

Now for CSV

```python
df1 = spark.read.csv("/Volumes/test/test_schema/test_volume/sales_corrupted.csv")
df1.display()
```

```python
schema_str = "id int, name string, product string, quantity int, price double, remarks string, _corrupt_record string"
df2 = spark.read.schema(schema_str).csv("/Volumes/test/test_schema/test_volume/sales_corrupted.csv")
df2.display()
```

**badRecordsPath:**

```python
df3 = spark.read.format('csv')\
.schema(schema_str)\
.option('badRecordsPath','/Volumes/test/test_schema/test_volume/bad_records')\
.load("/Volumes/test/test_schema/test_volume/sales_corrupted.csv")
df3.display()
```
