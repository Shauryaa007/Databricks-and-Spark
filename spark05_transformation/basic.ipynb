In a Dataframe data is stored parallelly so that it can sent partition data to diff executers

Every dataframe have some number of partitions (df.rdd.partitioned()) - not available in serverless computes

If You want to increase or decrease number of partitions to control

**Repartiotion -** 
In repartition we can either increase the number of partitions or we can decrese number of partitions but shuffling (data from any i/p partitions can go to any of the o/p partition) will be there.

**coalesce -**
It can only decrese partition and also avoid shuffling as it combine 2 or more partitions and results in a conolidated single partition.

```python
df2=df.repartition(2)
df2=df.coalesce(2)
```
